###  Matplotlib pointer
   - parameters : ms=20 or markersize=20
   - edge color : mec='b' or markeredgecolor='b'
   - face color : mfc='r' or markeredgecolor='r'

```python
y=[5,3,4,2,4]

plt.plot(y,'*--b',ms=30, mec='r', mfc='r')
```

### Matplotlib line

```python
y=[5,3,4,2,4]

plt.plot(y,'*--b',ms=30, mec='r', mfc='r', ls=':',c='g',lw=20)
```

### xlabel() ylabel()
        gán tên 2 trục
### Title()
        gán tên đồ thị
        
fontdict: font1

loc: 'left', 'right', 'center'


```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}


plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage",loc='center', fontdict=font2)
plt.title('Ngu ddaanf',fontdict=font1,loc='right')

plt.show()
```

### Scatter

- color: name, #RGB
- Each dot has a color: Create an array color and argument color=this_array

```python
x=np.array([1,4,5,4,3,4,5])
y=np.array([2,4,3,5,3,2,3])


plt.scatter(x,y,color='black')

x=np.array([3,5,4,8,9])
y=np.array([10,15,2,5,18])


plt.scatter(x,y,color='blue')

plt.show()
```

### color map: Bản đồ màu dùng trong scatter
Ex: color map: viridis(độ phủ màu từ 1->100)
- c : màu
- cmap: color map
- s: size of dots
- alpha:  trong suốt

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis',s=sizes,alpha =0.8)
plt.colorbar() # thêm bảng màu sang bên cạnh
plt.show()
```

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

### bar graph: biểu đồ cột
- color: name, RGB
- width(bar)
- height(barh)


```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y,color='green',width=0.5)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x,y,height=0.5)
plt.show()
```

### Pie chart: Biểu đồ tròn
- label
- startangle: góc bắt đầu  angle góc
- explode: Phần bùng nổ
- shadow: True or false
- color : name or RGB

#### Legend(): Chú thích
- title

```python
a=np.random.normal(100,10,(5))
labels=np.array(['DELL','ASUS','HP','ACER','MSI'])
explodes=np.array([0.2,0,0,0,0])

plt.pie(a,labels=labels, startangle=30,explode=explodes)
plt.legend(title='LAPTOP')

plt.show()
```
