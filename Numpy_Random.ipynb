### Phân bố Gauss

```python
x = random.normal(loc=1, scale=2, size=(2, 3))

print(x)
sns.distplot(x)
```

### Phân bố Nhị phân

```python
x = random.binomial(n=10, p=0.5, size=10)
print(x)
```

```python
sns.distplot(x)
```

```python
x = random.poisson(lam=2, size=10)
print(x)
sns.distplot(x)
```
